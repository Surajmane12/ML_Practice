```python
X = df.drop(columns = 'price',axis=1)
y = df['price']
X
```

**One whatt Encoding**

```python
X_one_encoded = pd.get_dummies(X,columns = ['model','transmission','fuelType'], drop_first=True)
X_one_encoded
```

```python
X_one_encoded = X_one_encoded.astype('int')
X_one_encoded
```

**LabelEncoding**

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
```

```python
columns = ['model','transmission','fuelType']
X_label_encode = X.copy()
X_label_encode
```

```python
label_encoders={}
for col in columns:
    le = LabelEncoder()
    X_label_encode[col] = le.fit_transform(X_label_encode[col].astype(str))
    label_encoders[col] = le
```

```python
X_label_encode
```

```python
from sklearn.preprocessing import StandardScaler
cols = ['mileage','tax','mpg','year','engineSize']
scaler = StandardScaler()

X_one_encoded[cols] = scaler.fit_transform(X_one_encoded[cols])
X_one_encoded
```

```python
X_label_encode.columns
```

```python

label_encode_cols = ['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize'] 

X_label_encode[label_encode_cols] = scaler.fit_transform(X_label_encode[label_encode_cols])

X_label_encode
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
 X_one_encoded, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]

adj_r2 = (1-(1-r2)*(n-1) / (n-p-1))  # formula
r2, adj_r2
```

```python
X_train, X_test, y_train, y_test = train_test_split(
 X_label_encode, y, test_size=0.33, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(X_train,y_train)
```

```python
y_pred = model2.predict(X_test)
y_pred
```
